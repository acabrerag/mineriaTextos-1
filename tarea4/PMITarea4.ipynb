Primero se eliminaron los signos de puntuación como las comas, dos puntos, punto, parentesis, etc. En bash.
Después se utilizó la libreria SnowballStemmer para buscar la raiz mediante el agoritmo de Porter.
Luego se buscaron todas las instancias de las palabras.
Después se hizo una matriz de frecuencias por documentos.
Y finalmente se utilizó la matriz de frecuencias para poder h

```python
import pandas as pd
import math as m
from nltk.stem.snowball import SnowballStemmer
import io
stemmer = SnowballStemmer("spanish")
```

```python
d = []#guardo las palabras en una lista de documentos
with open('corpus_DSM.txt') as f:
  for line in f:
    tok = line.split()
    for i in range(len(tok)-1):
        tok[i]=stemmer.stem(tok[i].decode('utf8')) #uso stemming en las palabras por Porter
    tok
    d.append(tok)
```

```python
elements=[]
numeros=[]
sumTotNumeros=0#total de palabras en el texto

for i in xrange(len(d)): 
    for j in xrange(len(d[i])):
        sumTotNumeros=sumTotNumeros+1
        if(d[i][j] not in elements):
            elements.append(d[i][j])
            numeros.append(1)
            #filas.append([d[i][j]])
            
        else:
            k=elements.index(d[i][j])
            numeros[k]=numeros[k]+1
            
```

```python
matrizOcurrenciasDoc=pd.DataFrame(0,index=elements,columns=xrange(len(d)))
```

```python
for i in xrange(len(d)):
    for j in d[i]:
        #if j in matrizOcurrenciasDoc.index:
            matrizOcurrenciasDoc.ix[j][i] = matrizOcurrenciasDoc.ix[j][i] + 1
        
```

```python
#Creo la el vector de probabilidades de cada elemento e inicializo la matriz  para observar las ocurrencias
probOcurrenciaTot=pd.Series(numeros,index=elements)
matrzPMI=pd.DataFrame(0,index=elements,columns=elements)
```

```python
for i in xrange (len(matrzPMI)):#tardó 15 minutos en correr
    for j in xrange (len(matrzPMI.ix[i])):
        for k in xrange (len(matrizOcurrenciasDoc.ix[0])):
            if(matrizOcurrenciasDoc.ix[i][k] >0 and matrizOcurrenciasDoc.ix[j][k]>0):
                matrzPMI.iloc[i,j]=matrzPMI.iloc[i,j]+1
```

```python
matrzPMI #matriz de todas las veces que aparecen juntos en los documentos
```

```python
matrizFinalPMI=matrzPMI
```

```python
sumTotNumeros=sumTotNumeros*1.0
for i in xrange (len(matrizFinalPMI)):
    for j in xrange (len(matrizFinalPMI.ix[i])):
        if(i ==j):
            matrizFinalPMI.iloc[i,j]=0
        else:
            if (matrizFinalPMI.iloc[i,j]!=0):
                matrizFinalPMI.iloc[i,j]=m.log10((matrzPMI.iloc[i,j]/sumTotNumeros)/((matrzPMI.iloc[i,i]+1.0/sumTotNumeros)*(matrzPMI.iloc[j,j]*1.0/sumTotNumeros)))
```
