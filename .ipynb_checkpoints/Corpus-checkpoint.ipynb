```python
import json
import numpy as np
import pandas as pd
from pprint import pprint

with open('Corpus.json') as data_file:    
    data = json.load(data_file)
```

```python
elements=[] #cuantos tags tenemos
numeros=[] #contador de iniciales
numeros2=numeros
print len(data["doc"]) #en total contamos con 333 documentos.
```

## Probabilidad de inicio

```python
#para cada documento si no está, lo adjunto a la lista de elements. Si está,  aumentamos el contador.
for i in range(len(data["doc"])-1): 
    if(data["doc"][i]["document"][0]["tag"] not in elements):
        elements.append(data["doc"][i]["document"][0]["tag"])
        numeros.append(1)
    else:
        j=elements.index(data["doc"][i]["document"][0]["tag"])
        numeros[j]=numeros[j]+1
        
        
```

```python
#para todas las ocurencias de las etiquetas, las sumo y divido entre la suma.
sumatot=sum(numeros)*1.0
for i in range(len(numeros)):
    numeros2[i]=numeros[i]/float(sumatot)
```

```python
#hago el df para las frecuencias iniciales.
frecuenciasIniciales=pd.Series(numeros2,index=elements)
```

```python
frecuenciasIniciales.to_csv("pi.csv", sep='|', encoding='utf-8')
```

```python
tokens=[]#cuales tokens hay en el corpus
tags=[]#cuales tags hay en el corpus
for i in xrange(len(data["doc"])-1):
    for l in range(len(data["doc"][i]["document"])-1):
        if(data["doc"][i]["document"][l]["token"] not in tokens):
            tokens.append(data["doc"][i]["document"][l]["token"])
for i in xrange(len(data["doc"])-1):
    for l in range(len(data["doc"][i]["document"])-1):
        if(data["doc"][i]["document"][l]["tag"] not in tags):
            tags.append(data["doc"][i]["document"][l]["tag"])
    
```
